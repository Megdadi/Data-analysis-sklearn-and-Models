# **plot**

```python
import matplotlib.pyplot as plt
#plt.style.use('Solarize_Light2')
#plt.style.use('dark_background')
#plt.style.use('bmh')
#plt.style.use('seaborn')
plt.style.use('classic')

x=[1,2,3,4,5]
y=[10,5,30,20,50]
y1=[18,15,55,30,20]
plt.plot(x,y, linewidth= 5, linestyle='dashed', color='r') # linestyle('solid or -', 'dashed or --', 'dashdot or -.', 'dotted or :')
plt.plot(x,y1, '-.g',linewidth= 5) # must write after x,y

plt.show()
```

```python
plt.style.use('seaborn')


area=[1000,1200,3300,4500,5000]
price1=[1044,1300,2000,1800,2500]
price2=[1800,1500,5500,3000,5000]
plt.plot(area,price1, linewidth= 5, linestyle='dashed', color='r', label= 'price1') # linestyle('solid or -', 'dashed or --', 'dashdot or -.', 'dotted or :')
plt.plot(area,price2, '-.g',linewidth= 5, label= 'price2') # must write after x,y
plt.title('area vs price')
plt.xlabel('area')
plt.ylabel('price')
plt.xlim(500,6000)
plt.ylim(1000,6000)
plt,plt.legend()
plt.show()
```

```python
plt.style.use('bmh')
area=[1000,1200,3300,4500,5000]
price1=[1044,1300,2000,1800,2500]
price2=[1800,1500,1000,3000,5000]
plt.subplot(2,1,1) # subplot(nrows, ncols, index)
plt.plot(area, price1)
plt.subplot(2,1,2) # subplot(nrows, ncols, index)
plt.plot(area, price2,':r')
```

# **scatter**

```python
plt.style.use('seaborn')
import numpy as np
x1=np.random.randint(10,100,40)
y1=np.random.randint(100,150,40)
y2=np.random.randint(100,150,40)


plt.scatter(x1, y1, s=30, c="r", marker=',') # marker='+', 'o', '<', 'x', 'd', '^', 'p', 'v', ','  ......  s= size
plt.scatter(x1, y2, s=30, c="g", marker='o') # marker='+', 'o', '<', 'x', 'd', '^', 'p', 'v', ','  ......  s= size

plt.show()
```

# **hist (one dimention)**

```python
a= np.array(['ahmad', 'ali', 'roro', 'yousef', 'mohammad', 'ebrahim', 'ahmad', 'roro', 'mohammad', 'roro'])
# alpha is the constract (0-1)
# histtype ('bar', 'barstacked', 'step', 'stepfilled' )
plt.hist(
    a,
    histtype='bar',
    alpha= 0.8,
    color= 'r',
    bins= 12
    
)
plt.show()
```

```python
a= np.array([['ahmad', 'ali', 'roro'],['yousef', 'mohammad', 'ebrahim'], ['ahmad', 'roro', 'mohammad', 'roro']])
# alpha is the constract (0-1)
# histtype ('bar', 'barstacked', 'step', 'stepfilled' )
plt.hist(
    a,
    histtype='bar',
    alpha= 0.8,
    color= ['r', 'b', 'g'],
    bins= 12
    
)
plt.show()
```

```python
import pandas as pd

dic={'day': ['satarday', 'sunday','monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'],
     'num_items': [3,5,1,7,4,2,9]}
df= pd.DataFrame(dic)
print(df)

plt.style.use('seaborn')
plt.plot(df['day'],df['num_items'], label='plot', alpha= 1)
plt.bar(df['day'],df['num_items'], edgecolor= 'black',color='g', label='bar',  alpha= 0.3)
plt.scatter(df['day'],df['num_items'], color='r', s=80, label='scatter', marker= 'x', alpha= 1)
plt.legend(loc='upper left')
plt.show()
```

# **PIE (one dim)**

```python
import matplotlib.pyplot as plt
cars= [3,5,2,1]
names= ['Audi', 'BMw', 'Mercidec', 'Ferrari']
plt.pie(cars, labels= names,
        labeldistance=1.1,
        radius=1.5,
        explode= [0.1, 0.01, 0.01, 0.01],
        colors= ['r', 'b', 'g', 'y']
)
```

# **plot3D**

```python
x=np.array([1,20,33,44,58])
y=np.array([80,180,170,100,150])
z=np.array([150,190,200,150,60])

x1= np.random.randint(8,100,50)
y1= np.random.randint(80,170,50)
z1= np.random.randint(100,160,50)

ax = plt.axes(projection='3d')
ax.set_xlim(1, 100)
ax.set_ylim(80, 200)
ax.set_zlim(100, 229)


ax.plot3D(x, y, z, color= 'r')
ax.scatter3D(x1, y1, z1, color='b')
```

# **seaborn**

```python
import pandas as pd
import seaborn as sns
d={'area':[500,560,1000,1500,2000,2500,3000,3500],
   'rooms':[4,3,6,7,3,8,5,5],
   'price':[4000,3500,6000,4500,5000,7000,6000,5500]
    
}
df= pd.DataFrame(d)
print(df.head(3),'\n\n')
corr_all_features=df.corr()  #relationship between each two features
corr_price=df.corrwith(df["price"])  #relationship between each features & target

print(corr_all_features,'\n\n')
print(corr_price.sort_values(ascending= False))
print('#################################### \n\n')
sns.heatmap(df.corr(), annot=True)
sns.pairplot(df)


```
