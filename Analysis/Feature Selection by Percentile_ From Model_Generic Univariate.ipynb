```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
cancer= load_breast_cancer()
X1= cancer.data
y1= cancer.target
print(X1.shape)
print(y1.shape)

X1= pd.DataFrame(X1)
y1= pd.DataFrame(y1)
data1= X1
data1.head(2)

```

```python
data1.columns= cancer.feature_names
```

```python
data1.head(2)
```

# **1. Feature Selection by Percentile (chi2 , f_classif) **

```python
from sklearn.feature_selection import SelectPercentile, chi2 , f_classif 
FeatureSelection = SelectPercentile(score_func = chi2, percentile=50) # score_func can be ( f_classif )
data= FeatureSelection.fit_transform(data1, y1)
print(data1.shape)
data1= pd.DataFrame(data1)
data1.head()
```

```python
releveant_features= FeatureSelection.get_support(indices=True)
releveant_features
```

```python
X1= X1.iloc[:,[ 0,  1,  2,  3,  6, 10, 12, 13, 20, 21, 22, 23, 25, 26, 27]]
X1.head(7)
```

# **2. Generic Univariate Select (chi2 , f_classif )**

```python
X2= cancer.data
y2= cancer.target
X2= pd.DataFrame(X2)
y2= pd.DataFrame(y2)
data2= X2
data2.columns= cancer.feature_names
data2.head(2)
```

```python
from sklearn.feature_selection import GenericUnivariateSelect, chi2, f_classif
selectfeature= GenericUnivariateSelect(score_func= f_classif,mode= 'k_best', param=15)
data2= selectfeature.fit_transform(data2,y2)
data2= pd.DataFrame(data2)
data2.head(2)
```

```python
releveant_features2=selectfeature.get_support(indices= True)
releveant_features2
```

```python
X2= X2.iloc[:,[ 0,  2,  3,  5,  6,  7, 10, 12, 13, 20, 22, 23, 25, 26, 27]]
X2.head(7)
```
