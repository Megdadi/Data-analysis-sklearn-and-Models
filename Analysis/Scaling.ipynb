```python
import pandas as pd
import numpy as np

dictionary =  {'A' : [1000, 200, 3, 14], 'B' : [400, np.nan, 2, 105], 'c' :[5000, 3,100, 1000], 'd' :['gh', 'jo', np.nan, 'dd']}
```

```python
dictionary
```

```python
df_dictionary= pd.DataFrame(dictionary)
df_dictionary
```

```python
########## get numeric data
df_dictionary = df_dictionary.select_dtypes(exclude=['object'])
# Or >>> df_dictionary = df_dictionary.select_dtypes(include=[np.number])

df_dictionary
```

```python
col=df_dictionary.columns
col
```

```python
######### Normalizer does not work with missing values
from sklearn.impute import SimpleImputer
imp= SimpleImputer(missing_values=np.nan, strategy='mean')
imp = imp.fit(df_dictionary)
norm_dictionary = imp.transform(df_dictionary)
norm_dictionary
```

```python
from sklearn.preprocessing import Normalizer 
scaler = Normalizer(norm='l2') # you can change the norm to 'l1' or 'max' 
scale_dictionary = scaler.fit_transform(norm_dictionary)
scale_dictionary
```

```python
scale_dictionary= pd.DataFrame(scale_dictionary)
scale_dictionary
```

```python
scale_dictionary.columns=df_dictionary.columns
scale_dictionary
```

```python
data=pd.read_csv('/content/drive/MyDrive/sklearn/cleaning/Marketting.csv')
```

```python
data.shape
```

```python
data= data.select_dtypes(include=[np.number]) # Normalizer does not work with missing values
print(data.shape)
data.head(3)
```

```python
data= data.drop(['id'], axis=1)
print(data.shape)
data.head(2)
```

```python
data= data[data.isnull().sum(axis=1)<10]  # remove all rows having nan's > 10
data= data[data.columns[data.isnull().sum() < 20]] # remove all coulmns having nan's > 20
print(data.shape)
```

```python
from sklearn.impute import SimpleImputer
imp= SimpleImputer (missing_values=np.nan, strategy='mean')
imp= imp.fit(data)
data_imp= imp.transform(data)
data_imp= pd.DataFrame(data_imp)
data_imp.head()
```

```python
'''
    from sklearn.preprocessing import StandardScaler 
    scaler = StandardScaler()
    scaler.fit(data)
    newdata = scaler.transform(data)
                                                   or
                                                      from sklearn.preprocessing import MinMaxScaler
                                                      scaler = MinMaxScaler()
                                                      scaler.fit(data)
                                                      newdata = scaler.transform(data)

''' 
from sklearn.preprocessing import Normalizer 
scaler = Normalizer(norm='l2') # you can change the norm to 'l1' or 'max' 
scale_data = scaler.fit_transform(data_imp)
scale_data
```

```python
scale_data= pd.DataFrame(scale_data)
scale_data.columns= data.columns
scale_data.head(3)
```
