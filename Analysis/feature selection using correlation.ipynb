# **Correlation**

```python
from scipy.stats import pearsonr
correlation=[]
i= 0
for i in range(len(features.columns)):
  corr,_= pearsonr(features.iloc[:,i], target.iloc[:,0]) # correlation between each column and target
  correlation.append(corr)

correlation
```

```python
correlation=pd.DataFrame(correlation)
correlation
```

```python
correlation=correlation.abs()
correlation
```

```python
releveant_features= []
i=0
for i in range (len(correlation)):
  corr_rel= correlation.iloc[i,0]
  if corr_rel > 0.5:
    releveant_features.append(i)   
releveant_features
```
