# **Categorical variables can be divided into two categories:**

1.    **Nominal: no particular order**
2.   **Ordinal: there is some order between values** 

```python
import numpy as np
import pandas as pd
```

# **Nominal**

**1.   One hot encoding**

**2.   Label Encoding**



```python
data = pd.DataFrame({ 'gender' : ['Male', 'Female', 'Male', 'Female', 'Female'],
                       'class' : ['A','B','C','D','A'],
                        'city' : ['Delhi','Gurugram','Delhi','Delhi','Gurugram'] })
data.head()
```

```python
# One hot encoding
encod_gender =pd.get_dummies(data['gender'])
encod_city= pd.get_dummies(data['city'])

# Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le_class = le.fit_transform(data[["class"]])
data['le_class']=le_class

new_data= pd.concat([data, encod_gender,encod_city],axis=1)
new_data

```

```python
new_data= new_data.drop(columns=['gender','city','class'])
new_data
```

# **Ordinal Encoding using dictionary**


```python
temp = {'temperature' :['very cold', 'cold', 'warm', 'hot', 'very hot']}
df=pd.DataFrame(temp)
print(df)
dictionary= {'very cold': 1,'cold': 2,'warm': 3,'hot': 4,"very hot":5}
df["temp_ordinal"] = df.temperature.map(dictionary)
print(df)
```
