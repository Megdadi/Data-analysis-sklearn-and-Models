# **train_test_split**

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
```

```python
Data= load_breast_cancer()
X= Data.data
y= Data.target
print(X.shape)

```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, random_state=44, shuffle=True)
```

```python
print('X_train shape is ' , X_train.shape)
print('X_test shape is ' , X_test.shape)
print('y_train shape is ' , y_train.shape)
print('y_test shape is ' , y_test.shape)
```

# **K_FOLDE**

```python
import numpy as np
from sklearn.model_selection import KFold

X = np.array([[1, 2], [3, 4], [5, 6], [7, 8]])
y = np.array([11, 22, 33, 44])

#nn = 2
#nn=3
nn=4

kf =  KFold(n_splits=nn, random_state=None, shuffle=False)
#kf.get_n_splits(X)
print(X)
print(y)
for train_index, test_index in kf.split(X):
    print("\n TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    print('X_train \n' , X_train)
    print('X_test \n' , X_test)
    print('y_train \n' ,y_train)
    print('y_test \n' , y_test)


```

# **Leave One Out**

```python
import numpy as np
from sklearn.model_selection import LeaveOneOut

X = np.array([1, 2, 3, 4])
y = np.array([5,6,7,8])

loo = LeaveOneOut()
#loo.get_n_splits(X)

LeaveOneOut()
for train_index, test_index in loo.split(X):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    print('X_train \n' , X_train)
    print('X_test \n' , X_test)
    print('y_train \n' ,y_train)
    print('y_test \n' , y_test)
```
