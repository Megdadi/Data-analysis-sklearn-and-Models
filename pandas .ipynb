# **indexing**

```python
d2 = {'a': [1,2,3,4],
       'b': [10,12,7,8],
       'c': [100,200,3,400],
       'd': [6,20,30,43],
       'e': [1,62,33,66]}
          
df2 = pd.DataFrame(d2)
print(df2,'\n')
print('##################################### \n', 'Indexing rows \n')
print(df2.iloc[0],'\n')
print(df2.iloc[[0]],'\n')
print(df2.iloc[[0, 1]],'\n')
print(df2.iloc[:2],'\n')
print('##################################### \n', 'Indexing both axes \n')
print(df2.iloc[0, 1],'\n')
print(df2.iloc[0:2,1:3],'\n')
print(df2.iloc[:,1:3],'\n')
print(df2.iloc[:-1],'\n') # all rows expect the last row
print(df2.iloc[:,-1],'\n') # last colmn
print('#########################################')
print(df2.columns)
```

# **insert , concat & dropping**

```python
d3={'col1':[1,3,5],'col2':[4,66,6], 'col3':[4,66,9]}
print(d3,'\n')
df4= pd.DataFrame(d3)
print(df4,'\n')
df4.insert(1, "newcol", [99, 99,99]) # insert(index, "name", [value1, value2])
print(df4,'\n')
df4= df4.drop('col1',axis=1) #drop column
print(df4,'\n \n')
df4= df4.drop(0,axis=0)# drop the first row
print(df4,'\n')
ls3=[[44,55],[66,77]]
se=pd.DataFrame(ls3,columns=['cl3','cl4'])
df5= pd.concat((df4,se),axis=1)
print(df5,'\n')
print(df5['cl4'])
print(df5['cl3'].describe())
```

# **Merge**

```python
l1=[['ahmad','writer'],
    ['ali','engineer'],
    ['yousef','doctor']]
l2= [['ali',60],
    ['yousef',40],
    ['ahmad',50]]
df_l1= pd.DataFrame(l1,columns=['name','jop'])
df_l2= pd.DataFrame(l2,columns=['name','age'])
print(df_l1,'\n \n',df_l2,'\n \n')
merg= pd.merge(df_l1,df_l2)
print(merg)
```

# **Groupping**

```python
ls4=[['engineering',500],
     ['acountant',800],
     ['officeboy',400],
     ['engineering',1000],
     ['HR',1500],
     ['acountant',900]]
df7= pd.DataFrame(ls4,columns=['section','salary'])     
print(df7,'\n\n')

print(df7.groupby('section').count(),'\n\n')
print(df7.groupby('section').sum(),'\n\n')
print(df7.groupby('section').mean())
```
