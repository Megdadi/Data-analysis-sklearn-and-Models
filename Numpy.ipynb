# **Indexing**

```python
x= np.random.randint(4,90,(4,5))
print(x,'\n')
print(x[:2],'\n')
print(x[2:],'\n')
print(x[-1],'\n')
print(x[:2],'\n')
print(x[1:3,2],'\n')
print(x[1:3],'\n')
print(x[3,4],'\n')
```

```python

ls=[[1,2,np.nan,4],[5,np.nan,7,8],[9,10,np.NAN,12],[3,16,np.NAN,22]]
arr_12=np.array(ls)
print(arr_12,'\n')
null_values_count= np.isnan(arr_12).sum()
print(null_values_count)
```

```python
arr_12[np.isnan(arr_12)] = 0
print(arr_12)
```

```python
arr_13= np.random.randint(6,90,(3,4))
print(arr_13,'\n')
# transpose
arr_13_T= np.transpose(arr_13)
print(arr_13_T,'\n')
arr_14= np.array([7,7,7])
arr_14= arr_14.reshape(3,1)
print(arr_14,'\n')
# appending
arr_appedent=np.append(arr_13,arr_14,axis=1)
print(arr_appedent)
```

```python
# How To Insert And Delete Array Elements?
arr_13= np.random.randint(6,90,(4,5))
print(arr_13,'\n')
print(len(arr_13),'\n') # number of rows
print(arr_13.size,'\n') 
```

# **How To Visualize NumPy Arrays**

---



```python
# Plotting an array with seaborn
import seaborn as sns
import matplotlib.pyplot as plt
x=np.random.randint(20,80,(4,2)) # create array
print(x)


sns.heatmap(x, annot=True, fmt="d") # annot=True to view numbers
plt.title(" array  using seaborn ",fontsize=12)
plt.show()
```

```python

import numpy as np 
from matplotlib import pyplot as plt 

x = np.arange(1,11) 
y = 2 * x + 5 
plt.title("Matplotlib demo") 
plt.xlabel("x axis caption") 
plt.ylabel("y axis caption") 
plt.plot(x,y,'o') 
plt.show()
```
